```python
def add_derivative(df, column, new_column_name):
    first_elem_idx = df.index.values[0]
    first_elem = df.loc[first_elem_idx, column]
    prev = df.shift(1)
    prev.loc[first_elem_idx, column] = first_elem
    new_column = (df[column] - prev[column]) / prev[column]
    df.loc[:, new_column_name] = new_column
    return df

def add_prev_value(df, column, new_column_name):
    first_elem_idx = df.index.values[0]
    first_elem = df.loc[first_elem_idx, column]
    prev = df.shift(1)
    prev.loc[first_elem_idx, column] = first_elem
    new_column = (df[column] - prev[column]) / prev[column]
    df.loc[:, new_column_name] = new_column
    return df
```

```python
def add_feature(df, df_der, column, new_column_name, type_='', window=3):
    if type_ == 'der':
        new_df = add_derivative(df_der, column, new_column_name)
    elif type_ == 'prev':
        new_df = df_der
        new_df.loc[:, new_column_name] = df_der[column]
    elif type_ == 'sma':
        new_df = df_der
        new_df.loc[:, new_column_name] = df_der.loc[:, column].rolling(window=window).mean()
    df.loc[:, new_column_name] = new_df[new_column_name]
    prev_el = np.nan
    for index, row in df.iterrows():
        if not np.isnan(row[new_column_name]):
            prev_el = row[new_column_name]
        else:
            df.loc[index, new_column_name] = prev_el
    df.loc[:, new_column_name] = df[new_column_name]#.shift(1)
    return df
```

```python
def gen_feat(df, column):
    for day_week in ['tue', 'thu', 'sat', 'sun']:
        for type_ in ['der', 'prev', 'sma']:
            if type_ == 'sma':
                add_feature(df, df[df['is_' + day_week] == 1], column, 
                            day_week + '_' + type_ + '3', type_=type_, window=3)
                add_feature(df, df[df['is_' + day_week] == 1], column, 
                            day_week + '_' + type_ + '10', type_=type_, window=10)
            else:
                add_feature(df, df[df['is_' + day_week] == 1], column, day_week + '_' + type_, type_=type_)

# new_df = initialize_df()
# gen_feat(new_df)
```

```python
errors = []
sma_errors = []
cum_errors = []
# all_errors = []
# all_sma_errors = []
# all_cum_errors = []
for i, category_id in enumerate(set(df['category_id'])):
    print('cat_id: ', category_id, 'i: ', i)
    new_df = initialize_df(category_id)
    new_df.loc[:, 'y'] = new_df['quantity'] - new_df['quantity'].shift(1)

    gen_features(new_df, 'y', ['sma'])

    new_df = new_df.fillna(0.)
    new_df = reduce_features(new_df, 'sma3')
    add_prev_values(new_df, 'y', 12)
    new_df = new_df.fillna(0.)
    
    columns_to_select = ['y' + str(i) for i in range(1, 13)] + ['shipping_date', 'quantity', 'sma3', 'y']
    learn_df = new_df.loc[:, columns_to_select]
    learn_df = learn_df.loc[13:, :]

    df_train, df_test = learn_df[learn_df['shipping_date'] < '2019-11-15'], \
                        learn_df[learn_df['shipping_date'] >= '2019-11-15']
    
    y_train, y_test = np.array(df_train['y']), np.array(df_test['y'])
    X_train , X_test = np.array(df_train.drop(columns=['y', 'shipping_date', 'quantity', 'sma3'])), \
                       np.array(df_test.drop(columns=['y', 'shipping_date', 'quantity', 'sma3']))
    if len(y_train) > 0 and len(y_test) > 0:
    
        clf = Ridge(alpha=1.0)
        clf.fit(X_train, y_train)

        err = rel_error(df_test['quantity'][1:], clf.predict(X_test)[1:] + df_test['quantity'].shift(1)[1:])
        if abs(err) < 0.5:
            errors.append(err)
            err = rel_error(df_test['quantity'][1:], df_test['quantity'][1:] + df_test['sma3'][1:])
            sma_errors.append(err)
            pred = (clf.predict(X_test)[1:] + df_test['quantity'].shift(1)[1:]) * 0.3 + (df_test['quantity'][1:] + df_test['sma3'][1:]) * 0.7
            err = rel_error(df_test['quantity'][1:], pred)
            cum_errors.append(err)
    clear_output()
```
